```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

val = pd.read_csv('~/Downloads/tmpjs7xlh5z.csv')
rotation_angles = pd.read_csv('~/smartspim-registration/rotation_angles.csv')

val = val.merge(rotation_angles, on='subject_id')

fig, ax = plt.subplots(nrows=3, ncols=1, figsize=(15, 15))
sns.scatterplot(val, x=val['AP_rotation'].abs(), y='rmse', ax=ax[0])
sns.scatterplot(val, x=val['SI_rotation'].abs(), y='rmse', ax=ax[1])
sns.scatterplot(val, x=val['ML_rotation'].abs(), y='rmse', ax=ax[2])
plt.show()

pd.DataFrame({
    'AP_rotation_abs': val['AP_rotation'].abs(),
    'SI_rotation_abs': val['SI_rotation'].abs(),
    'ML_rotation_abs': val['ML_rotation'].abs(),
    'rmse': val['rmse']
}).to_csv('/tmp/val_rotation.csv', index=False)
```

```python
g = sns.relplot(
    data=val,
    x="slice_idx",
    y="rmse",
    col="subject_id",
    col_wrap=4,
    kind="scatter",
    height=3,
    aspect=1.2
)
plt.show()
```
